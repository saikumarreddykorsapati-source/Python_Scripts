```python
import pandas as pd
from tkinter import *
from tkinter import filedialog
import os
import sys
from datetime import datetime
time = datetime.now()

yes_list = []
no_list = []
error_in_function = []
mreview_error = []

def browseFiles(title_arg,file_type1,file_ext1):
    filepath = filedialog.askopenfilename(initialdir = os.getcwd()+"/assets/", title = title_arg,filetypes = ((file_type1,file_ext1), ("all files","*.*")))
    if filepath=="":
        sys.exit("\nNo file selected from opened pop up window")
    filename = os.path.basename(filepath).split("/")[-1]
    ext = filename.split(".")
    for i in ext:
       ext = filename.split(".")[-1]

#     print("\nChoosen Input File Path : " + filepath)
#     print("\t\n--> Choosen Input File Name : " + filename,end="\n\n")
#     print("\t--> Original File Format : "+ ext,end="\n\n")
    return filepath
window = Tk()
button_explore = Button(window,text = "Browse Files", command = browseFiles)
window.withdraw()  

errors_list = []
errors_Excel_filePath = browseFiles("Select Error's listed Excel File to Load Error keywords to check","xlxs files", ".*xlsx")
x = errors_Excel_filePath.split("/")[-1]
y = x.split(".")[-2]


server_log_filePath = browseFiles("Select Server Log File","server log files", ".*log")
df = pd.read_excel(errors_Excel_filePath)
for i in range(len(df)):
    errors_list.append([df['ERRORs List'].values[i]])
# print("ERROR's list : ",errors_list,"\n\n")

with open(server_log_filePath) as file:
    contents = file.read()
    for j in errors_list:
        for i in j:
            if i.lower() in contents.lower():
#                 print (i.upper() ,': \tYES - Word Found')
                yes_list.append(i+"  YES - Word Found")
            else:
#                 print (i.upper() ,':\tno not found')
                no_list.append(i+"  NO - Word NOT Found")
                
    
            if ("ERROR IN FUNCTION - "+i).lower() in contents.lower():
                print("ERROR IN FUNCTION - "+i)
                error_in_function.append("ERROR IN FUNCTION - "+i)
                
            if ("MREVIEW."+i).lower() in contents.lower():
                print("MREVIEW."+i)
                mreview_error.append("MREVIEW."+i)
            
# print(yes_list)
df1 = pd.DataFrame(yes_list)
df2 = pd.DataFrame(no_list)
df3 = pd.DataFrame(error_in_function)
df4 = pd.DataFrame(mreview_error)

# print("Yes Error's List =",df1,"\n\n\n")
# print("No Error's List =",df2)

with pd.ExcelWriter('C:\\Users\\894226\\Downloads\\checked_'+y+"__"+time.strftime("%d-%b-%Y")+'.xlsx') as writer:
   
    # use to_excel function and specify the sheet_name and index
    # to store the dataframe in specified sheet
    df1.to_excel(writer, sheet_name="YES_ERRORS_Found", index=False)
    df2.to_excel(writer, sheet_name="NO_ERRORS_NOT_Found", index=False)
    df3.to_excel(writer, sheet_name="error_in_function", index=False)
    df4.to_excel(writer, sheet_name="mreview_error", index=False)

```
